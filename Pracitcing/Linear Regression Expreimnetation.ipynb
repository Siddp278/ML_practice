```python
# plot for residual error 
  
## setting plot style 
plt.style.use('fivethirtyeight') 
  
## plotting residual errors in training data 
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, 
            color = "green", s = 10, label = 'Train data') 
  
## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 3) 
  
## plotting legend 
plt.legend(loc = 'upper right') 
  
## plot title 
plt.title("Residual errors") 
  
## function to show plot 
plt.show()
```

```python
#Using Least Square Method to eliminate bs.
X_array = np.array(X_train)
X_Transpose = X_array.transpose()
```

```python
Arr_mul = np.matmul(X_Transpose,X_array)
Arr_mul
```

```python
Arr_inverse = np.linalg.inv(Arr_mul)
Arr_mul1 = np.matmul(Arr_inverse,X_Transpose)
Arr_mul2 = np.matmul(Arr_mul1,y_train)
#Till here we have found out beta(cap).
```
